````python
import pandas as pd 
import numpy as np
````

````python
df = pd.read_csv('origindata/premier-league-matches.csv')

conditions = [
    df['HomeGoals'] == df['AwayGoals'],                         # ë¬´ìŠ¹ë¶€
    (df['FTR'] == 'H') & (df['HomeGoals'] > df['AwayGoals']),   # í™ˆ ìŠ¹
    (df['FTR'] == 'A') & (df['AwayGoals'] > df['HomeGoals'])    # ì›ì • ìŠ¹
]

# ê° ì¡°ê±´ì— ëŒ€ì‘í•˜ëŠ” ê°’
choices = [1, 3, 3]

# ì¡°ê±´ì„ ë§Œì¡±í•˜ì§€ ì•ŠëŠ” ê²½ìš°(íŒ¨ë°°)ëŠ” 0
df['Points'] = np.select(conditions, choices, default=0)

# --- 3) Points ê¸°ì¤€ ì •ë ¬(ë‚´ë¦¼ì°¨ìˆœ) ---
df = df.sort_values(by='Points', ascending=False)

print(df)
````

````python
output_path = 'dataset/results_with_points.csv' 
df.to_csv(output_path, index=False, encoding='utf-8-sig')
````

````python
# 2) í™ˆ/ì›ì • ê°ê° ìŠ¹ì  ê³„ì‚° â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def points_per_side(row):
    if row['HomeGoals'] > row['AwayGoals']:      # í™ˆ ìŠ¹
        return pd.Series({'HomePts': 3, 'AwayPts': 0})
    elif row['HomeGoals'] < row['AwayGoals']:    # ì›ì • ìŠ¹
        return pd.Series({'HomePts': 0, 'AwayPts': 3})
    else:                                        # ë¬´ìŠ¹ë¶€
        return pd.Series({'HomePts': 1, 'AwayPts': 1})

df[['HomePts', 'AwayPts']] = df.apply(points_per_side, axis=1)

# 3) í™ˆÂ·ì›ì • ì •ë³´ë¥¼ í•˜ë‚˜ë¡œ ë¬¶ì–´ íŒ€â€‘ì‹œì¦Œë³„ë¡œ ì§‘ê³„ â”€â”€â”€â”€â”€â”€â”€â”€
home = (
    df[['Season_End_Year', 'Home', 'HomePts']]
      .rename(columns={'Home': 'Team', 'HomePts': 'Points'})
)
away = (
    df[['Season_End_Year', 'Away', 'AwayPts']]
      .rename(columns={'Away': 'Team', 'AwayPts': 'Points'})
)

points_long = pd.concat([home, away], ignore_index=True)

team_season_pts = (
    points_long
      .groupby(['Team', 'Season_End_Year'], as_index=False)['Points']
      .sum()
)

# 4) í”¼ë²—: í–‰=Team, ì—´=Season_end_year â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
pivot = (
    team_season_pts
      .pivot(index='Team', columns='Season_End_Year', values='Points')
)

# 5) ë¹ ì§„ ì‹œì¦Œ(ì—´) 0ìœ¼ë¡œ ì±„ìš°ê³  ì •ë ¬ â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
all_seasons = list(range(1992, 2026))            # 1992â€†~â€†2025

pivot = (
    pivot
      .reindex(columns=all_seasons)   
      .fillna(0)                      
      .astype(int)                    
      .sort_index(axis=1)             
      .sort_index()                   
)

print(pivot.head(5))
pivot.to_csv('dataset/season_points_by_team.csv')
````

````python
df2 = pd.read_csv('dataset/season_points_by_team.csv')

df2['sum'] = df2.loc[:, '1992':'2025'].sum(axis=1)

top10 = (
    df2.sort_values(by='sum', ascending=False)
       .reset_index()
       .loc[:, ['Team', 'sum']]
       .rename(columns={'Team': 'íŒ€ ì´ë¦„', 'sum': 'ì´ ìŠ¹ì '})
)

top10.index += 1  # ìˆœìœ„ 1ë¶€í„° ì‹œì‘í•˜ë„ë¡
print(top10)
````

## 2. ì—°ë„ë³„ Eloâ€¯Ranking ìˆœìœ„ ì‚°ì¶œ & ë¹„êµ

### 2â€‘1. ì—°ë„ë³„ ì „ì²´ ìˆœìœ„ í…Œì´ë¸” 
elo_rank = (
    elo_pivot
      .rank(axis=0,                # ì—´(=ì—°ë„)ë§ˆë‹¤
            method='min',          # ê³µë™ 1ìœ„ â†’ ë‘˜ ë‹¤ 1
            ascending=False)       # Eloê°€ í´ìˆ˜ë¡ 1ìœ„
      .astype('Int64')             # NaN í—ˆìš© ì •ìˆ˜í˜•
)

>  **ê²°ê³¼ ì˜ˆì‹œ (ì¼ë¶€)**  
> | club | 2000 | 2001 | â€¦ | 2024 | 2025 |
> |------|------|------|---|------|------|
> | Arsenal | 3 |  2 | â€¦ | 2| 2 |
> | Tottenham | 8 | 7 | â€¦ | 9 | 6|


````python
elo_df = pd.read_csv('origindata/EloRatings.csv')

elo_df['year'] = pd.to_datetime(elo_df['date']).dt.year

# 3) í´ëŸ½â€‘ì—°ë„ë³„ Elo í•©ê³„ ê³„ì‚° â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
club_year_elo = (
    elo_df.groupby(['club', 'year'], as_index=False)['elo']
      .sum()                                 
)


elo_pivot = (
    club_year_elo
      .pivot(index='club', columns='year', values='elo')
      .fillna(0)    
      .sort_index()                  # í´ëŸ½ëª… ì•ŒíŒŒë²³ìˆœ
      .sort_index(axis=1)            # ì—°ë„ ì˜¤ë¦„ì°¨ìˆœ
)

elo_pivot.to_csv('dataset/elo_by_club_year.csv', encoding='utf-8-sig')

print(elo_pivot.head(5))
````

````python
# 1) ì—°ë„ë³„ í´ëŸ½ Elo ìˆœìœ„ ê³„ì‚° (1ìœ„ = 1)
elo_rank = elo_pivot.rank(axis=0,      # ì—´(=ì—°ë„)ë³„ë¡œ
                          method='min',# ê³µë™ 1ìœ„ë©´ ë‘˜ ë‹¤ 1
                          ascending=False).astype('Int64')

# 2) íŠ¹ì • ì—°ë„(ì˜ˆ: 2023) Topâ€‘10 ë³´ê¸°
year = 2023
top10_2023 = (
    elo_pivot[year]                 # ì—°ë„ë³„ Elo ì ìˆ˜ ì‹œë¦¬ì¦ˆ
      .nlargest(10)                 # ìƒìœ„ 10ê°œ
      .reset_index()
      .rename(columns={year: 'Elo'})
)
top10_2023['Rank'] = range(1, 11)
print(top10_2023)

# 3) ëª¨ë“  ì—°ë„ì— ëŒ€í•´ Topâ€‘N ì¶”ì¶œ (dict í˜•íƒœ)
def top_n_by_year(pivot, n=5):
    out = {}
    for y in pivot.columns:
        top_n = pivot[y].nlargest(n).reset_index()
        top_n['Rank'] = range(1, n+1)
        out[y] = top_n
    return out

top5_each_year = top_n_by_year(elo_pivot, n=5)

# ì˜ˆì‹œ ì¶œë ¥: 2000ë…„ ìƒìœ„ 5íŒ€
print(top5_each_year[2000])
````
